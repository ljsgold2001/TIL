```python
import copy

def rotate(key):
    changedkey = []
    keysize = len(key)
    
    changedkey = [[0 for col in range(keysize)] for row in range(keysize)]
    
    for i in range(keysize):
        for j in range(keysize):
            changedkey[i][j] = key[keysize-1-j][i]
    return changedkey
    

def check(key , Map):
    keysize = len(key)
    locksize = len(Map) - (keysize-1)*2
    validmove = len(Map) -(keysize-1)
    
    
    for i in range(0,validmove):
        for j in range(0,validmove):
            count = 0
            temp = copy.deepcopy(Map)
            for  row in range(i,i+keysize):
                for col in range(j,j+keysize):
                    temp[row][col] += key[row - i][col -j]
            '''print("======================")
            for k in temp:
                print(k , sep = "\n")'''
            
            for row in range(keysize-1, validmove):
                for col in range(keysize-1, validmove):
                    if(temp[row][col] !=1):
                        break;
                    count +=1
            if (count == locksize *locksize):
                return True
    return False
                
                

def solution(key, lock):
    M = len(key)
    N = len(lock)
    
    map_N = N + (M-1) * 2
    Map = [[0 for i in range(map_N)]  for j in range(map_N)]
    #for i in Map:
        #print(i,sep = "\n")
    for i in range(M, M+N):
        for j in range(M, M+N):
            Map[i-1][j-1] = lock[i-M][j-M]
    #for i in Map:
        #print(i,sep = "\n")
    
    for i in range(0,4):
        checkresult = check(key, Map)
        if checkresult == True:
            return True
        else:
            key = rotate(key)
    return False
```
