```python
def bfs( n , maps):
    dx =[0,0,1,-1]
    dy =[1,-1,0,0]
    global count
    
    while q:
        
        for i in range(len(q)):
            x,y = q.popleft()
            visited[x][y] = True
            print(count)
            print(x,y)
            for i in range(4):
                nx = x +dx[i]
                ny = y +dy[i]
                print("@@@@@@",nx,ny)
                if(0 <= nx and nx < n and 0 <= ny and ny < n and maps[nx][ny] !=0 and visited[nx][ny] ==False):
                    print("@@@" , nx,ny)
                    maps[nx][ny] = count
                    q.append([nx,ny])
                    
        
        print("@@")
        print(count)
        print(maps)
        print(visited)
        
    
```

```python
from collections import deque
n = int(input())


maps = [] #입력받을 map 생성후에 입력받기
visited = [[False]*n for _ in range(n)] #visited 리스트 생성
q = deque()
count = 1


for _ in range(n):
    tmp = [int(i) for i in input()]
    maps.append(tmp)

for i in range(n):
    for j in range(n):
        if maps[i][j] ==1 and visited[i][j] == False:
      
            q.append([i,j])
        

bfs(n,maps)
            
```
